```python
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import Chroma
from langchain_core.documents import Document 
```

```python
doc1 = Document(
        page_content="Virat Kohli is an Indian cricketer known for his aggressive batting style. "
                     "He has scored over 25,000 international runs and is one of the best chase masters.",
        metadata={"name": "Virat Kohli", "country": "India", "role": "Batsman", "format": "All"}
    )
doc2 = Document(
        page_content="MS Dhoni is a former Indian captain famous for his finishing ability and calm personality. "
                     "He led India to ICC World Cup 2011 and Champions Trophy 2013 victories.",
        metadata={"name": "MS Dhoni", "country": "India", "role": "Wicket-Keeper", "format": "ODI/T20"}
    ) 
doc3 = Document(
        page_content="Ben Stokes is an English all-rounder known for clutch performances. "
                     "He played a key role in England's 2019 World Cup win.",
        metadata={"name": "Ben Stokes", "country": "England", "role": "All-Rounder", "format": "Test/ODI"}
    )
doc4 = Document(
        page_content="Kane Williamson is a technically strong batsman from New Zealand. "
                     "He is respected for his leadership and calm demeanor.",
        metadata={"name": "Kane Williamson", "country": "New Zealand", "role": "Batsman", "format": "Test/ODI"}
    )
doc5 = Document(
        page_content="Jasprit Bumrah is one of the world's best fast bowlers known for yorkers and accuracy. "
                     "He plays a crucial role in India's bowling attack.",
        metadata={"name": "Jasprit Bumrah", "country": "India", "role": "Bowler", "format": "All"}
    )
```

```python
docs = [doc1, doc2, doc3, doc4, doc5]
```

```python
vector_store = Chroma(
    embedding_function = OpenAIEmbeddings(),
    persist_directory = "my_chroma_db",
    collection_name = "sample"
)
```

```python
vector_store.add_documents(docs)
```

```python
vector_store.get(include= ['embeddings','documents', 'metadatas'])
```

```python
vector_store.similarity_search(
    query = "who among these are from India",
    k = 3 # the number of doc we want as a result
)
```

```python
vector_store.similarity_search_with_score(
    query = "who among these are from India",
    k = 3 # the number of doc we want as a result
)
```

```python
vector_store.similarity_search(
    query = "",
    filter = {'country' : 'India'}
)
```
